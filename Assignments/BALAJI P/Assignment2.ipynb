## 4. Handle the Missing Values

```python
df.dropna(axis=0, inplace=True)
```

## 5. Find the outliers and replace the outliers


```python
df
```

```python
from scipy import stats

z = np.abs(stats.zscore(df[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]))
print(np.where(z > 3))
```

```python
df = df[(z < 3).all(axis=1)]
df
```

## 5. Encoding Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df['Geography'] = le.fit_transform(df['Geography'])
df['Gender'] = le.fit_transform(df['Gender'])
df
```

## 6. Split the data into dependent and independent variables

```python
indep_var = df.iloc[:,0:10].values
indep_var
```

```python
dep_var = df.iloc[:,10:].values
dep_var
```

## 7. Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split  
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(indep_var, dep_var, test_size=0.3, random_state=0)
```
